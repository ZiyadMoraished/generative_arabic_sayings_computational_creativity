```python
import requests
from bs4 import BeautifulSoup
```

# collect links for all cats links

```python
# https://xn--sgb8bg.net/%d8%ad%d9%83%d9%85-%d9%88-%d8%a3%d9%85%d8%ab%d8%a7%d9%84/
with open('hekm.net_links.txt', 'r') as f: 
    html = f.read() 
    
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
quote_titles = soup.find_all(class_='post-title')
quote_titles = [title.text.strip() for title in quote_titles]

quote_links = soup.find_all('a')
quote_links = [tag.attrs for tag in quote_links]
quote_links = [list(tag.values())[0] for tag in quote_links]
```

# Loop over each author page to get their sayings

```python
for link, title in zip(quote_links, quote_titles):
    req = requests.get(link, verify=False)
    html = req.content
    soup = BeautifulSoup(html,'html.parser')
    if 'شعر' in title:
        print()
        print(title, '-->', link)
        print()
        continue
    # if an author has multiple pages
# #     try:
#     try:
#         pagination = soup.find(class_='pagination justify-content-center').find_all('li')
#         pages_count = pagination[-1].text.strip()
#         pages_count = int(pages_count)
#     except Exception as e:
#         print(complete_link, '-->', str(e))
    # if author has a single page
#     except:
#         pages_count = 1
        
    # loop over th pages    
#     for author_page in range(1, pages_count+1):
#         print(author_page)
    all_quotes = soup.find_all(class_='quotescollection-quote')
    quotes_text_list = [text.find('p').text.strip() for text in all_quotes]
#     print(quotes_text_list)

#         print(quote_list)
    # write all sayings to a text file
    with open("hekm.net_sayings.txt", "a") as f:
        for quote  in quotes_text_list:
#                 print(quote)
            try:
                quote = quote.strip().replace('\n',' ')
            except Exception as e:
                
                print('Error:', link, str(e))
                
            f.write('\n{}'.format(quote))
```
